```python
print(len(name))
print(len(price))
print(len(review))
```

```python

min_length = min(len(name), len(price), len(review))

name = name[:min_length]
price = price[:min_length]
review = review[:min_length]

print(len(name))
print(len(price))
print(len(review))
```

```python
df=pd.DataFrame({"Product Name ":name ,"Product Price":price ,"Review":review})
df['Product Price']=df['Product Price'].str.replace('â‚¹','')
```

```python
print(df.tail())
print(df.to_csv('final.csv'))
print(df)
```
